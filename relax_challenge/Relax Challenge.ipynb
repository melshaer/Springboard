```python
import pandas as pd
import matplotlib.pyplot as plt
```

The data are available in two tables: a table containing information about subscribed product users and another containing product usage dates and times for individual users.

```python
with open('takehome_users.csv') as f:
    print(f.encoding)
    
with open('takehome_user_engagement.csv') as f:
    print(f.encoding)
```

```python
users_df = pd.read_csv('takehome_users.csv', encoding='cp1252')
usage_df = pd.read_csv('takehome_user_engagement.csv', encoding='cp1252')
```

```python
users_df.head()
```

```python
usage_df.head()
```

```python
#Merge both tables
merged_df = pd.merge(left=users_df, right=usage_df, left_on='object_id', right_on='user_id').drop(columns='object_id')
#merged_df.head()
merged_df.sort_values(by='time_stamp', inplace=True)
merged_df.tail(20)
```

```python
max_date = max(merged_df['time_stamp'])
min_date = min(merged_df['time_stamp'])

merged_df['time_stamp'] = pd.to_datetime(merged_df['time_stamp'])

merged_df['day'] = merged_df['time_stamp'].dt.day
merged_df['month'] = merged_df['time_stamp'].dt.month
merged_df['year'] = merged_df['time_stamp'].dt.year
merged_df['week'] = merged_df['time_stamp'].dt.week
merged_df.groupby('year')['visited'].count()
```

It is clear from the merged table that the year "2013" had the most user activity. It would be interesting to find the number of users per day in a week.

```python
num_users_per_day_week = merged_df.groupby(['year', 'week', 'day'])['user_id'].count()
num_users_per_day_week.reindex()
num_users_per_day_week[(2012, 52)].plot()
plt.title('Number of users in week 52, year 2012')
plt.show()
```

```python
total_activity_per_week = merged_df.groupby(['year', 'week'])['visited'].sum()
total_activity_per_week.reindex()
total_activity_per_week[2012].plot()
plt.title('Total Activity in 2012')
plt.show()
```

```python
num_users_per_day_week[(2012, 22)].plot()
plt.title('Number of users in week 22, year 2012')
plt.show()
```

```python
total_activity_per_week[2013].plot()
plt.title('Total Activity in 2013')
plt.show()
```

```python
total_activity_per_week[2014].plot()
plt.title('Total Activity in 2014')
plt.show()
```

```python
total_activity_per_day = merged_df.groupby(['year', 'week', 'day'])['visited'].sum()
total_activity_per_day.reindex()
total_activity_per_day[2012].plot()
plt.title('Total Activity in 2012')
plt.show()
```

Defining an adopted user as a user that has login activity on three different days in at least one week period. We are interested to explore the following question:

What factors affect user adoption?

```python
#Filter adopted users
g = merged_df.groupby(['user_id', 'year', 'month', 'week', 'day']).sum()
v = merged_df.groupby(['user_id', 'year', 'month', 'week'])['visited'].sum()
d = g.loc[v == 3]
d
#merged_df[(merged_df['user_id']==2) & (merged_df['year']==2013)]
```

If the user used the product more than once in a day, it's counted only once.

Factor Analysis and Prediction Model

Using Principal Components Analysis, we could represent our data using components or factors that best explain the variance in the data. We could then use the generative model to predict new user information or behavior that fit the patterns in our original data.

```python
from sklearn.decomposition import PCA

X = d.reset_index().values

pca = PCA()
X_new = pca.fit_transform(X)
pca.explained_variance_ratio_
```
